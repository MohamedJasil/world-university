### Q. Which is the Top 10 universities in the year 2013?

```python
univ_df["total_score"]=pd.to_numeric(univ_df["total_score"])
univ_df[(univ_df["total_score"]>=85) & (univ_df["year"] ==2013)].head(10)
```

#### We can see that there are some unknown values in the dataset. lets ignore it.

```python
univ_df=univ_df[univ_df["total_score"]!="-"]
univ_df["total_score"]=pd.to_numeric(univ_df["total_score"])
top = univ_df[(univ_df["total_score"]>=85) & (univ_df["year"] ==2013)].head(10)
top
```
